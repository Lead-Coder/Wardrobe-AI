```python
import json
import csv
import os
import pandas as pd  # Import pandas

# Directory containing the JSON files
directory_path = "/kaggle/input/fashion-product-images-dataset/fashion-dataset/styles/"

# Fields to extract
fieldnames = [
    "productID", "productDisplayName", "ageGroup", "gender", "baseColour", "colour1", "colour2",
    "fashionType", "season", "year", "usage", "displayCategories",
    "Pattern", "typeName", "description", "TopOrBottom", "Fit", "Sleeve Length", "Neck", "Fabric", "brandName"
]

# Helper function to safely extract values from normalized data
def get_value(df, key):
    try:
        # Handling both direct access and cases where the field might be missing
        return df[key].iat[0] if key in df.columns else None
    except Exception as e:
        print(f"Error extracting {key}: {e}")
        return None

# Open the CSV file for writing
with open('all_extr.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    count = 0

    for filename in os.listdir(directory_path):
        #print(f"Processing file {count+1} of 10: {filename}")
        count += 1
        if filename.endswith('.json'):
            file_path = os.path.join(directory_path, filename)

            try:
                # Extract product ID from filename (remove .json)
                product_id = filename.split('.')[0]

                # Open the file and load the data
                with open(file_path, 'r', encoding='utf-8') as file:
                    data = json.load(file).get('data', [])

                if data:  # Ensure data exists
                    # Normalize JSON data using pandas
                    data_normalized = pd.json_normalize(data, sep='.')

                    # Prepare the row to be written to CSV
                    row = {
                        "productID": product_id,
                        "productDisplayName": get_value(data_normalized, "productDisplayName"),
                        "ageGroup": get_value(data_normalized, "ageGroup"),
                        "gender": get_value(data_normalized, "gender"),
                        "baseColour": get_value(data_normalized, "baseColour"),
                        "colour1": get_value(data_normalized, "colour1"),
                        "colour2": get_value(data_normalized, "colour2"),
                        "fashionType": get_value(data_normalized, "fashionType"),
                        "season": get_value(data_normalized, "season"),
                        "year": get_value(data_normalized, "year"),
                        "usage": get_value(data_normalized, "usage"),
                        "displayCategories": get_value(data_normalized, "displayCategories"),
                        "Pattern": get_value(data_normalized, "articleAttributes.Pattern"),
                        "typeName": get_value(data_normalized, "articleType.typeName"),
                        "description": get_value(data_normalized, "productDescriptors.description.value"),
                        "TopOrBottom": get_value(data_normalized, "subCategory.typeName"),
                        "Fit": get_value(data_normalized, "articleAttributes.Fit"),
                        "Sleeve Length": get_value(data_normalized, "articleAttributes.Sleeve Length"),
                        "Neck": get_value(data_normalized, "articleAttributes.Neck"),
                        "Fabric": get_value(data_normalized, "articleAttributes.Fabric"),
                        "brandName": get_value(data_normalized, "brandName")
                    }

                    # Write the row to the CSV
                    writer.writerow(row)
                else:
                    print(f"No 'data' found in {filename}")

            except Exception as e:
                print(f"Failed to process {filename}: {e}")
```
